```python
import re
from pdfminer.high_level import extract_text,extract_pages
```

```python
text = extract_text("cvsamples.pdf")
print(text)
```

```python
for page in extract_pages("file-sample_150kB.pdf"):
    for element in page:
        print(element)
```

```python
print(text)
```

```python
email = re.compile(r"[a-zA-Z0-9]+@[a-zA-Z]+\.[a-zA-Z]+")

matches = email.findall(text)
print(set(matches))
```

```python
education = re.compile(r"(Bachelor of Science|Bsc|PhD|Doctor of Philosophy|MA|Msc) in ([A-Za-z ]+)")
education_matcher = education.findall(text)
```

```python
print(education_matcher)
```

```python
header = re.compile(r"[A-Z]+ [A-Z]*")
header_text = header.findall(text)
print(set(header_text))
```

```python
Experience = re.compile(r"(Experience |experience )?(more than |more )?[0-9]+(Years |year |years |Years )?(Experience |experience )?[A-Za-z]*")
Experience_matcher = Experience.findall(text)
print(set(Experience_matcher))
```
